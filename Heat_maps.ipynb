```python
#a big thanks to David Hoffman for showing me this library and some of basics

from folium import plugins
from folium.plugins import HeatMap
import pandas as pd
import folium

train_df = pd.read_csv('./input/train.csv')

#Where in world and we focusing our map
map_heat = folium.Map(location=[train_df.Latitude.mean(), train_df.Longitude.mean()], 
                   zoom_start = 10.3)

# Ensure you're handing it floats
train_df['Latitude'] = train_df['Latitude'].astype(float)
train_df['Longitude'] = train_df['Longitude'].astype(float)

train_df.Date = pd.to_datetime(train_df.Date, format='%Y-%m-%d')
train_df['Year'] = train_df.Date.map(lambda x: x.year)

#creating Dataframes were are going to use to map heatmaps by Year
heat_df_07 = train_df[(train_df['Year']==2007) & (train_df['WnvPresent']==1)]
heat_df_09 = train_df[(train_df['Year']==2009) & (train_df['WnvPresent']==1)]
heat_df_11 = train_df[(train_df['Year']==2011) & (train_df['WnvPresent']==1)] 
heat_df_13 = train_df[(train_df['Year']==2011) & (train_df['WnvPresent']==1)]


def plot_heat(df):
    # List comprehension to make out list of lists, this is for 2007
    heat_data = [[row['Latitude'],row['Longitude']] for index, row in df[['Latitude', 'Longitude']].iterrows()]
    
    # Plot it on the map
    HeatMap(heat_data).add_to(map_heat)
    
    # Display the map
    map_heat
```

```python
plot_heat(heat_df_07)
map_heat
```

```python
plot_heat(heat_df_09)
map_heat
```

```python
plot_heat(heat_df_11)
map_heat
```

```python
plot_heat(heat_df_13)
map_heat
```
